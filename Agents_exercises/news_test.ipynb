```python
import requests
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
# default language & country = all 
def top_headlines():
    url = (f"https://newsapi.org/v2/top-headlines?language=en&sortBy=popularity&apiKey={os.environ["NEWS_API_ORG_KEY"]}")
    response = requests.get(url).json()["articles"]
    return response
```

```python
def topic_news(query):
    url = (f"https://newsapi.org/v2/everything?q={query}&apiKey={os.environ["NEWS_API_ORG_KEY"]}")
    response = requests.get(url).json()["articles"]
    return response
```

```python
result = topic_news(query="Data Engineering")
```
